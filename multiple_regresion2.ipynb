```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
```

```python
#seaborn
import seaborn as sns
sns.regplot(x=df['R&D Spend'],y=y,color='red')
```

```python
#one-hot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

col_transform = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), ['State']), remainder='passthrough')
x=col_transform.fit_transform(x)
x
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# Create a dictionary with the data
data = {'Actual': y_test, 'Predicted': y_pred,'error': y_test-y_pred}

# Create a dataframe from the dictionary
df1 = pd.DataFrame(data)

# Print the dataframe
print(df1)
```

```python
# Print the intercept and slope
print("Intercept is", model.intercept_)
print('slopes is')
list(zip(x,model.coef_))

```

```python
#performance evaluation

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score

# Calculate the evaluation metrics
metrics = {
    'Mean Absolute Error': mean_absolute_error(y_test, y_pred),
    'Mean Absolute Percentage Error': mean_absolute_percentage_error(y_test, y_pred),
    'Mean Squared Error': mean_squared_error(y_test, y_pred),
    'Root Mean Squared Error': np.sqrt(mean_squared_error(y_test, y_pred)),
    'R2 Score': r2_score(y_test, y_pred)
}

# Print the combined metrics
for metric_name, metric_value in metrics.items():
    print(f'{metric_name}: {metric_value}')
```
